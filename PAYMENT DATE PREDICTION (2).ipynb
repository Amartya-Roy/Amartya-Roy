# Read the data

```python
import pandas as pd 
data = pd.read_csv(r"C:\Users\nEW u\Desktop\all shortcuts\COURSERA MACHINE LEARNING BY ANDREW Ng\CSV FILES\1607232_Full_Data.csv")
data.head(50)
```

# Remove the isopen = 1 

```python
data.drop(data[data.Isopen==1].index, axis=0, inplace=True)
```

# Removing the constant columns

```python
for col in data:
    if len(data[col].unique())==1:
        data.drop(col, axis=1, inplace=True)
        
```

```python
data
```

```python
data.dtypes
```

# Feature selection

```python
data.drop(['Company Code', 'Fiscal Year', 'Payment Terms','Job Id','Document Date','Customer Name'], axis=1, inplace=True)
data
```

# Sorting the values

```python
data.sort_values('Due Date Norm')
data
```

# Date parsing

```python
import datetime
data['Clearing_Date_Norm_dt'] = pd.to_datetime(data['Clearing Date Norm'])
data['Clearing_Date_Norm_Day']=data['Clearing_Date_Norm_dt'].dt.day
data['Clearing_Date_Norm_Month']=data['Clearing_Date_Norm_dt'].dt.month
data['Clearing_Date_Norm_Year']=data['Clearing_Date_Norm_dt'].dt.year
data
```

```python
import datetime
data['Document_Date_Norm_dt'] = pd.to_datetime(data['Document Date Norm'])
data['Document_Date_Norm_Day']=data['Document_Date_Norm_dt'].dt.day
data['Document_Date_Norm_Month']=data['Document_Date_Norm_dt'].dt.month
data
```

```python
import datetime
data['Due_Date_Norm_dt'] = pd.to_datetime(data['Due Date Norm'],errors ='coerce')
data['Due_Date_Norm_Day']=data['Due_Date_Norm_dt'].dt.day
data['Due_Date_Norm_Month']=data['Due_Date_Norm_dt'].dt.month
data['Due_Date_Norm_Year']=data['Due_Date_Norm_dt'].dt.year
data
```

```python
import datetime
data['Baseline_Date_Norm_dt'] = pd.to_datetime(data['Baseline Date Norm'],errors ='coerce')
data['Baseline_Date_Norm_Day']=data['Baseline_Date_Norm_dt'].dt.day
data['Baseline_Date_Norm_Month']=data['Baseline_Date_Norm_dt'].dt.month
data
```

# Delete some similar columns

```python
data.drop(["Document Date Norm", "Baseline Date Norm","Due Date Norm","Clearing Date Norm","Due Date"], axis = 1, inplace = True)
data
```

# Inserting the delay column(y)

```python
data['Delay']=(data['Clearing_Date_Norm_dt']-data['Due_Date_Norm_dt']).dt.days
data.dtypes
```

# Creating  new features

## 1

```python
import numpy as np
data['DelayIndicator'] = np.where(data['Delay']<=0, '1', '-1')
```

## 2

```python
data['payment clearing gap']=(data['Document_Date_Norm_Day']-data['Clearing_Date_Norm_Day'])
data['Payment Clearing Indicator'] = np.where((data['payment clearing gap']<=10) | (data['payment clearing gap']<0), '-1', '1')
data
```

## 3

```python
data['Base - Due']=(data['Baseline_Date_Norm_dt']-data['Due_Date_Norm_dt']).dt.days
data['Doc - Due']=(data['Document_Date_Norm_dt']-data['Due_Date_Norm_dt']).dt.days
data
```

# Label encoding

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
data['Currency']= label_encoder.fit_transform(data['Currency'])
data['Currency'].unique() 
data
```

# Preparation of training-testing

```python
data.drop(['Clearing_Date_Norm_dt','Document_Date_Norm_dt','Baseline_Date_Norm_dt'],axis = 1,inplace = True)
data
```

```python
data.dtypes
```

# Parsing float64 to int64

```python
data['Delay'] = data['Delay'].fillna(0).astype(np.int64)
data['Base - Due'] = data['Base - Due'].fillna(0).astype(np.int64)
data['Doc - Due'] = data['Doc - Due'].fillna(0).astype(np.int64)
data['Due_Date_Norm_Day'] = data['Due_Date_Norm_Day'].fillna(0).astype(np.int64)
data['Due_Date_Norm_Month'] = data['Due_Date_Norm_Month'].fillna(0).astype(np.int64)
data['Due_Date_Norm_Year'] = data['Due_Date_Norm_Year'].fillna(0).astype(np.int64)
data['Baseline_Date_Norm_Day'] = data['Baseline_Date_Norm_Day'].fillna(0).astype(np.int64)
data['Baseline_Date_Norm_Month'] = data['Baseline_Date_Norm_Month'].fillna(0).astype(np.int64)
data.isnull().sum()
```

# Splitting Into Train&Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data.drop('Delay',axis=1), data['Delay'], test_size=0.2)
X_train.drop(['payment clearing gap'],axis = 1,inplace = True)
X_test.drop(['payment clearing gap'],axis = 1,inplace = True)
X_train.drop(['Due_Date_Norm_dt'],axis = 1,inplace = True)
X_test.drop(['Due_Date_Norm_dt'],axis = 1,inplace = True)
X_train.drop(['Clearing_Date_Norm_Day'],axis = 1,inplace = True)
X_test.drop(['Clearing_Date_Norm_Day'],axis = 1,inplace = True)
X_train.drop(['Clearing_Date_Norm_Month'],axis = 1,inplace = True)
X_test.drop(['Clearing_Date_Norm_Month'],axis = 1,inplace = True)
X_train.drop(['Clearing_Date_Norm_Year'],axis = 1,inplace = True)
X_test.drop(['Clearing_Date_Norm_Year'],axis = 1,inplace = True)
X_train.drop(['Document_Date_Norm_Day'],axis = 1,inplace = True)
X_test.drop(['Document_Date_Norm_Day'],axis = 1,inplace = True)
X_train.drop(['Document_Date_Norm_Month'],axis = 1,inplace = True)
X_test.drop(['Document_Date_Norm_Month'],axis = 1,inplace = True)
X_train.drop(['Baseline_Date_Norm_Day'],axis = 1,inplace = True)
X_test.drop(['Baseline_Date_Norm_Day'],axis = 1,inplace = True)
X_train.drop(['Baseline_Date_Norm_Month'],axis = 1,inplace = True)
X_test.drop(['Baseline_Date_Norm_Month'],axis = 1,inplace = True)
X_train.drop(['Payment Clearing Indicator'],axis = 1,inplace = True)
X_test.drop(['Payment Clearing Indicator'],axis = 1,inplace = True)
X_train.drop(['Actual Open Amount'],axis = 1,inplace = True)
X_test.drop(['Actual Open Amount'],axis = 1,inplace = True)
X_train.drop(['Doc - Due'],axis = 1,inplace = True)
X_test.drop(['Doc - Due'],axis = 1,inplace = True)
X_train.drop(['Document Id'],axis = 1,inplace = True)
X_test.drop(['Document Id'],axis = 1,inplace = True)

```

# Model Estimation




## 1. Applying Random Forest Regressor Model 

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10,max_depth =10, random_state = 0)
regressor.fit(X_train,Y_train)
y_pred = regressor.predict(X_test)
y_pred
```

## Accuracy

###### Test_score

```python
accuracy = regressor.score(X_test,Y_test)
accuracy
```

##### Train_score

```python
accuracy = regressor.score(X_train,Y_train)
accuracy
```

## 2. Applying Decision Tree Regressor Model

```python
from sklearn.tree import DecisionTreeRegressor
regressor1 = DecisionTreeRegressor(random_state = 0,max_depth = 20)
regressor1.fit(X_train, Y_train)
y_pred1 = regressor1.predict(X_test)
y_pred1
```

## Accuracy 

#### Test_score

```python
accuracy = regressor1.score(X_test,Y_test)
accuracy
```

#### Train_score

```python
accuracy = regressor1.score(X_train,Y_train)
accuracy
```
