```python
import pandas as pd 
data = pd.read_csv(r"C:\Users\nEW u\Desktop\all shortcuts\COURSERA MACHINE LEARNING BY ANDREW Ng\CSV FILES\1607232_Full_Data.csv")
data.head(50)
```

```python
data.columns
```

```python
gk = data.groupby('Fk Customer Map Id') 
gk.first()
```

```python
data.drop(data[data.Isopen==1].index, axis=0, inplace=True)
```

```python
for col in data:
    if len(data[col].unique())==1:
        data.drop(col, axis=1, inplace=True)
```

```python
data.drop(['Company Code', 'Fiscal Year','Job Id','Document Date','Customer Name'], axis=1, inplace=True)
data
```

```python
data.sort_values('Due Date Norm')
data
```

```python
import datetime
data['Document_Date_Norm_dt'] = pd.to_datetime(data['Document Date Norm'])
data['Due_Date_Norm_dt'] = pd.to_datetime(data['Due Date Norm'],errors ='coerce')
data['Baseline_Date_Norm_dt'] = pd.to_datetime(data['Baseline Date Norm'],errors ='coerce')
data['Clearing_Date_Norm_dt'] = pd.to_datetime(data['Clearing Date Norm'])
```

```python
data.drop(["Document Date Norm", "Baseline Date Norm","Due Date Norm","Clearing Date Norm","Due Date"], axis = 1, inplace = True)
data
```

```python
data['Delay']=(data['Clearing_Date_Norm_dt']-data['Due_Date_Norm_dt']).dt.days
data
```

```python
import numpy as np
data['DelayIndicator'] = np.where(data['Delay']<=0, '1', '-1')
data
```

```python
data['Base - Due']=(data['Baseline_Date_Norm_dt']-data['Due_Date_Norm_dt']).dt.days
data['Doc - Due']=(data['Document_Date_Norm_dt']-data['Due_Date_Norm_dt']).dt.days
data['Base - Doc']=(data['Baseline_Date_Norm_dt']-data['Document_Date_Norm_dt']).dt.days
data
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
data['Currency']= label_encoder.fit_transform(data['Currency'])
data['Currency'].unique() 
data
```

```python
data.drop(['Clearing_Date_Norm_dt'],axis = 1,inplace = True)
data
```

```python
data.dtypes
```

```python
data['Delay'] = data['Delay'].fillna(0).astype(np.int64)
data['Base - Due'] = data['Base - Due'].fillna(0).astype(np.int64)
data['Doc - Due'] = data['Doc - Due'].fillna(0).astype(np.int64)
data['Base - Doc'] = data['Base - Doc'].fillna(0).astype(np.int64)
data.dtypes
```

```python
data
```

```python
 data.rename(index=str, columns={"Payment Terms": "Pmt"},inplace = True)
data
```

```python
data['Payment_Numeric']=data.Pmt.str.extract('(\d+)')
data
```

```python
data['Payment_Numeric'] = data['Payment_Numeric'].fillna(0).astype(np.int64)
```

```python
import numpy as np
data['Current Dispute Amount Indicator'] = np.where(data['Current Dispute Amount']==0, '1', '-1')
data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data.drop('Delay',axis=1), data['Delay'], test_size=0.2)
X_train.drop(['Due_Date_Norm_dt'],axis = 1,inplace = True)
X_test.drop(['Due_Date_Norm_dt'],axis = 1,inplace = True)
X_train.drop(['Baseline_Date_Norm_dt'],axis = 1,inplace = True)
X_test.drop(['Baseline_Date_Norm_dt'],axis = 1,inplace = True)
X_train.drop(['Document_Date_Norm_dt'],axis = 1,inplace = True)
X_test.drop(['Document_Date_Norm_dt'],axis = 1,inplace = True)
X_train.drop(['Pmt'],axis = 1,inplace = True)
X_test.drop(['Pmt'],axis = 1,inplace = True)
X_train.drop(['Current Dispute Amount'],axis = 1,inplace = True)
X_test.drop(['Current Dispute Amount'],axis = 1,inplace = True)
X_train.drop(['Open Amount Norm'],axis = 1,inplace = True)
X_test.drop(['Open Amount Norm'],axis = 1,inplace = True)
X_train.drop(['Document Id'],axis = 1,inplace = True)
X_test.drop(['Document Id'],axis = 1,inplace = True)
X_train.drop(['Base - Due'],axis = 1,inplace = True)
X_test.drop(['Base - Due'],axis = 1,inplace = True)
X_train.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators =10,max_depth =10, random_state = 0)
regressor.fit(X_train,Y_train)
y_pred = regressor.predict(X_test)
y_pred
```

```python
accuracy = regressor.score(X_test,Y_test)
accuracy
```

```python
accuracy = regressor.score(X_train,Y_train)
accuracy
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor1 = DecisionTreeRegressor(random_state = 0,max_depth = 20)
regressor1.fit(X_train, Y_train)
y_pred1 = regressor1.predict(X_test)
y_pred1
```

```python
accuracy = regressor1.score(X_test,Y_test)
accuracy
```

```python
accuracy = regressor1.score(X_train,Y_train)
accuracy
```
